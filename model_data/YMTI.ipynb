* 음식 카테고리 분류

양식(W), 중식(C), 일식(J), 한식(K), 디저트(D), 기타(E)

```python
# df_edit[df_edit['favor_food'].str.contains('77')]

# target_string = ['19','20','21']
# df_edit[df_edit['favor_food'].map(lambda x: all(string in x for string in target_string))]
```

```python
food_cat=[]
for i in range(len(df_edit)):
    food_cat.append( df_edit['favor_food'][i].split(','))
    del food_cat[i][-1]

# del food2[-1]
food_cat[:5]
```

```python
for i in range(len(food_cat)):
    for j in range(len(food_cat[i])):
        # 양식
        for a in range(1,8):
            if food_cat[i][j] == str(a):
                food_cat[i][j] = 'western'
        for c in range(11,14):
            if food_cat[i][j] == str(c):
                food_cat[i][j] = 'western' 
        for l in range(80,82):
            if food_cat[i][j] == str(l):
                food_cat[i][j] = 'western'    
        for o in range(89,92):
            if food_cat[i][j] == str(o):
                food_cat[i][j] = 'western' 
        for q in range(96,99):
            if food_cat[i][j] == str(q):
                food_cat[i][j] = 'western'
        if food_cat[i][j] == '100':
            food_cat[i][j] = 'western'
        
        # 디저트
        for b in range(8,11):
            if food_cat[i][j] == str(b):
                food_cat[i][j] = 'dessert'
        for d in range(14,16):
            if food_cat[i][j] == str(d):
                food_cat[i][j] = 'dessert'   
        for e in range(17,19):
            if food_cat[i][j] == str(e):
                food_cat[i][j] = 'dessert'          
        if food_cat[i][j] == '99':
            food_cat[i][j] = 'dessert'

         # 일식
        if food_cat[i][j] == '16':
            food_cat[i][j] = 'japanese'   
        for f in range(19,32):
            if food_cat[i][j] == str(f):
                food_cat[i][j] = 'japanese'

        # 중식
        for g in range(32,47):
            if food_cat[i][j] == str(g):
                food_cat[i][j] = 'chinese'
                
        # 한식
        for h in range(47,72):
            if food_cat[i][j] == str(h):
                food_cat[i][j] = 'korean' 
        for n in range(84,89):
            if food_cat[i][j] == str(n):
                food_cat[i][j] = 'korean'
        for p in range(92,96):
            if food_cat[i][j] == str(p):
                food_cat[i][j] = 'korean'

        # 기타(동남아 등)
        for k in range(72,80):
            if food_cat[i][j] == str(k):
                food_cat[i][j] = 'etc'
        for m in range(82,84):
            if food_cat[i][j] == str(m):
                food_cat[i][j] = 'etc'
 


food_cat[:8]
```

```python
# 리스트 중복제거
new_food_cat =[]

for i in range(len(food_cat)):
    new_food_cat.append(list(set(food_cat[i])))
    # result = ','.join(new_food_cat)
    

new_food_cat[:8]

```
